```python
import math
```

```python
fuelsum = 0
with open('inputs/1.txt', 'r') as file:
    content = file.readlines()
    for line in content:
        print(line)
        fuelsum += math.floor(int(line)/3)-2
    
print(fuelsum)
```

```python
def calc_fuel(fuel, fuelsum):
    cfuel = math.floor(fuel/3)-2
    fuelsum += cfuel
    if cfuel >= 0:
        return calc_fuel(cfuel, fuelsum)
    else:
        return fuelsum
    
```

```python
fuelsum = 0
for line in content:
    fuelsum += calc_fuel(int(line), 0)
```
